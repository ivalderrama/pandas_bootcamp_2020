### Filtering dataframes with many conditions (AND)

```python
import pandas as pd
```

```python
titanic = pd.read_csv('../Course_Materials_Part1/Video_Lecture_NBs/titanic.csv')
```

```python
titanic.head(10)
```

```python
mask1 = titanic['sex'] == 'male'
mask1
```

```python
mask2 = titanic['age'] > 14
mask2
```

```python
(mask1 & mask2)
```

```python
male_surv = titanic.loc[mask1 & mask2, ['survived', 'pclass', 'sex', 'age']]
male_surv.head(20)
```

```python
male_surv.info()
```

```python
male_surv.describe()
```

```python
titanic.describe()
```

### Filtering dataframes with many conditions (OR)

```python
titanic
```

```python
mask1 = titanic.sex == 'female'
mask1.head(20)
```

```python
mask2 = titanic['age'] < 14
mask2.head(20)
```

```python
(mask1 | mask2).head(20)
```

```python
titanic.loc[mask1 | mask2]
```

```python
wom_or_child = titanic.loc[mask1 | mask2, ['survived', 'pclass', 'sex', 'age']]
wom_or_child.head(20)
```

```python
wom_or_child.info()
```

```python
wom_or_child.describe()
```

```python
titanic.describe()
```
