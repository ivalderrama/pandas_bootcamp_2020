### Advanced indexing and slicing

```python
import pandas as pd
```

```python
summer = pd.read_csv('../Course_Materials_Part1/Video_Lecture_NBs/summer.csv')
```

```python
summer
```

#### Case 1: Getting the 1st 5 rows and rows 354 and 765

```python
list(range(5))
```

```python
list(range(5)) + [354, 765]
```

```python
rows = list(range(5)) + [354, 765]
```

```python
summer.iloc[rows]
```

#### Case 2: Getting the 1st 3 columns and the columns 'Gender' and 'Event'

```python
summer.columns
```

```python
summer.columns[:3]
```

```python
summer.columns[:3].to_list()
```

```python
summer.columns[:3].to_list() + ['Gender', 'Event']
```

```python
col = summer.columns[:3].to_list() + ['Gender', 'Event']
```

```python
summer.loc[:, col]
```

#### Case 3: Combining position and label based indexing
##### Rows at positions 200 and 300 and columns "Athlete" and "Medal"

```python
summer
```

```python
summer.loc[[200, 300], ["Athlete", "Medal"]]
```

#### Case 4: Combining position and label based indexing
##### Rows 'PHELPS Michael' and position columns 4 and 6

```python
summer = pd.read_csv('../Course_Materials_Part1/Video_Lecture_NBs/summer.csv', index_col='Athlete')
```

```python
summer
```

```python
summer.columns
```

```python
summer.columns[[4, 6]]
```

```python
col = summer.columns[[4, 6]]
```

```python
summer.loc['PHELPS, Michael', col]
```
