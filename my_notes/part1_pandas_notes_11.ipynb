### Renaming index and column labels

```python
import pandas as pd
```

```python
summer = pd.read_csv("../Course_Materials_Part1/Video_Lecture_NBs/summer.csv", index_col="Athlete")
```

```python
summer
```

```python
summer.index[0]
```

```python
summer.index[0] = 'HAJOS, Alfred'

# Error
# Index is immutable, unless a list is passed 
```

```python
# This is the work around

summer.rename(mapper = {"HAJOS, Alfred":'HAYOS, Alfred'}, axis="index")
```

```python
summer.rename(index = {"HAJOS, Alfred":'HAYOS, Alfred'}, inplace=True)
```

```python
summer
```

```python
summer.rename(mapper = {"Sex":"Gender", "City":"Host_City"}, axis = "columns")
```

```python
summer
```

```python
summer.rename(columns = {"Sex":"Gender", "City":"Host_City"}, inplace=True)
```

```python
summer
```

### Filtering dataframes with on condition

```python
titanic = pd.read_csv('../Course_Materials_Part1/Video_Lecture_NBs/titanic.csv')
```

```python
titanic
```

```python
titanic['sex']
```

```python
pd.options.display.max_rows = 900

titanic['sex'] == 'male'
```

```python
titanic[titanic['sex'] == 'male']
```

```python
titanic.loc[titanic['sex'] == 'male']
```

```python
titanic.loc[titanic['sex'] == 'male', 'fare']
```

```python
# For readability can do the following:

mask1 = titanic['sex'] == 'male'
mask1
```

```python
titanic_male = titanic.loc[mask1]
```

```python
titanic_male.head()
```

```python
titanic.dtypes
```

```python
titanic.dtypes == object
```

```python
mask2 = titanic.dtypes == object
mask2
```

```python
titanic.loc[:, ~mask2]  # Note: ~ (tilda) means NOT
```

```python
titanic.loc[mask1, ~mask2]  # Note: ~ (tilda) means NOT
```
