### Indexing and  slicing with reindex()

```python
import pandas as pd
```

```python
summer = pd.read_csv('../Course_Materials_Part1/Video_Lecture_NBs/summer.csv')
```

```python
summer
```

```python
summer.loc[[0, 5, 30000], ['Athlete', 'Medal']]
```

```python
summer.reindex(index=[0, 5, 30000, 40000], columns=['Athlete', 'Medal', 'Age'])
```

```python
summer = pd.read_csv('../Course_Materials_Part1/Video_Lecture_NBs/summer.csv', index_col='Athlete')
```

```python
summer
```

```python
summer.reindex(columns=['Medal', 'Age'])
```

```python
summer.reindex(index=['PHELPS, Michael'], columns=['Medal', 'Age'])

# Error
# Can't reindex with duplicates
```
